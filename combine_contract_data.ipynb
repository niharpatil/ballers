```python
import pandas as pd
import numpy as np
import re
from os import listdir
from os.path import isfile, join
print(pd.__version__)
```

```python
df = pd.read_excel('./data/orig_data/Free Agent Contract Data 2007 Raw.xlsx')
```

```python
def processContractDataset(datapath, file):
    df = pd.read_excel(datapath + file)
    
    # Set Column headers and drop duplicate rows
    df.columns = np.array(df.iloc[0])
    df = df.drop([0],axis=0)
    df = df.drop_duplicates()
    
    # Change year of signing date to reflect actual signing date
    year = int(file[25:29])
    df['Signed'] = year
    
    # Populate NaN values for players not offered contracts
    df['Dollars'] = df['Dollars'].fillna(0)
    df['Years'] = df['Years'].fillna(0)
    
    # Create feature representing average salary per years
    df['Dollars_Per_Year'] = df['Dollars'] / df['Years']
    df['Dollars_Per_Year'] = df['Dollars_Per_Year'].fillna(0)
        
    df = df.drop(columns=['Status','From','To','Years','Pos', 'Dollars'])
    
    return df
```

```python
# Grab all files concerning free agent contract data
datapath = './data/orig_data/'
onlyfiles = [f for f in listdir(datapath) if isfile(join(datapath, f))]

# Datasets with "Contract" in the filename
contractfiles = [k for k in onlyfiles if 'Contract' in k]

# Combine all contract datasets into one dataframe
total_df = pd.DataFrame()

for file in contractfiles:
    df = processContractDataset(datapath, file)
    total_df = total_df.append(df)
```

```python
# Write out combine contract datasets to file

total_df.to_csv(path_or_buf='./data/computed/total_contract_data.csv')
```
